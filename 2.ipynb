```python
import torch
from torch import nn
from torch.autograd import Variable
import torch.nn.functional as F
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
```

```python
features, labels = load_breast_cancer(return_X_y=True)
```

```python
class Model(nn.Module):
    def __init__(self, input_dim):
        super(Model, self).__init__()
        self.linear1 = nn.Linear(input_dim,30)
        self.linear2 = nn.Linear(30, 100)
        self.linear3 = nn.Linear(100, 2)
    def forward(self, x):
        x - F.relu(self.linear1(x))
        x = F.relu(self.linear2(x))
        x = F.softmax(self.linear3(x), dim=0)
        return x
```

```python
model = Model(features.shape[1])
```

```python
loss_fn = nn.NLLLoss()
optimizer = torch.optim.Adam(model.parameters())
```

```python
features_train, features_test,labels_train, labels_test = train_test_split(features, labels, shuffle=True, random_state=34)
x_train = Variable(torch.from_numpy(features_train), requires_grad=False).float()
y_train = Variable(torch.from_numpy(labels_train), requires_grad=False).long()

x_test = Variable(torch.from_numpy(features_test), requires_grad=False).float()
y_test = Variable(torch.from_numpy(labels_test), requires_grad=False).long()
```

```python
def train(epochs):
    for epoch in range(epochs):
        print ("Train Epoch #",epoch)
        y_pred = model(x_train)
        loss = loss_fn(y_pred, y_train)
        print ("Train Loss: ", loss.item())
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
```

```python
def test(epochs):
    with torch.no_grad():
        for epoch in range(epochs):
            print ("Test Epoch #", epoch)
            y_pred = model(x_test)
            loss = loss_fn(y_pred, y_test)
            print ("Test loss:", loss.item())
        
```

```python
# Training
train(100)
```
